```python

# selected_columns = ['Year', 'Month', 'Mean Temp (°C)']
# df_selected = df[selected_columns]

# # Map month numbers to month names
# month_names = {
#     1: 'JAN', 2: 'FEB', 3: 'MAR', 4: 'APR', 5: 'MAY', 6: 'JUN',
#     7: 'JUL', 8: 'AUG', 9: 'SEP', 10: 'OCT', 11: 'NOV', 12: 'DEC'
# }
# df_selected['Month'] = df_selected['Month'].map(month_names)

# # Pivot the dataframe
# df_pivot = df_selected.pivot(index='Year', columns='Month', values='Mean Temp (°C)')

# # Print the transformed dataset
# df_pivot




selected_columns = ['Year', 'Month', 'Mean Temp (°C)']
df_selected = df[selected_columns]

# Map month numbers to month names
month_names = {
    1: 'JAN', 2: 'FEB', 3: 'MAR', 4: 'APR', 5: 'MAY', 6: 'JUN',
    7: 'JUL', 8: 'AUG', 9: 'SEP', 10: 'OCT', 11: 'NOV', 12: 'DEC'
}
df_selected['Month'] = df_selected['Month'].map(month_names)

# Pivot the dataframe
df_pivot = df_selected.pivot(index='Year', columns='Month', values='Mean Temp (°C)')

# Add the 'Year' column back to the DataFrame
df_pivot['Year'] = df_pivot.index

# Reorder the columns
columns_order = ['Year', 'JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']
df_result_Transform = df_pivot[columns_order]

# Print the transformed dataset
df_result_Transform


```

```python
# Export the transformed dataset to a CSV file
df_result_Transform.to_csv('Woodbridge_Temp_Data_Transformed.csv', index=False)
```

```python
# Drop duplicated rows
df_result_Transform_Clean = df_result_Transform.drop_duplicates()

# Drop rows with missing values
df_result_Transform_Clean = df_result_Transform.dropna()
```

```python
df_result_Transform_Clean
```

```python

min_year = df_result_Transform_Clean['Year'].min()
max_year = df_result_Transform_Clean['Year'].max()

# Create a range of years from min_year to max_year (inclusive)
all_years = range(int(min_year), int(max_year) + 1)

# Get the unique years from the 'Year' column
existing_years = df_result_Transform_Clean['Year'].unique()

# Find the missing years by subtracting the existing years from the complete range
missing_years = set(all_years) - set(existing_years)

# Print the missing years
print("Missing years:")
for year in missing_years:
    print(year)

```

```python
# Export the transformed dataset to a CSV file
df_result_Transform_Clean.to_csv('Woodbridge_Temp_Data_Transformed_Clean.csv', index=False)
```
