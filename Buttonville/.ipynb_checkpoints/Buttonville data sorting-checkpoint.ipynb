```python
import pandas as pd
```

```python
path = 'en_climate_monthly_ON_615HMAK_1986-2015_P1M.csv'
```

```python
df = pd.read_csv(path)
```

```python
df.head()
```

```python
df['Month'] = df['Month'].apply(lambda x: pd.to_datetime(str(x), format='%m').strftime('%b'))
```

```python
df_temp = df.pivot(index='Year', columns='Month', values='Mean Temp (°C)')
```

```python
months_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
df_temp = df_temp[months_order]
```

```python
df_temp = df_temp.reset_index().rename_axis(None, axis=1)
```

```python
df_temp.head()
```

```python
df_temp.to_csv("Mean Temperature (1986-2015).csv")
```

```python
df_prec = df.pivot(index='Year', columns='Month', values='Total Precip (mm)')
```

```python
df_prec = df_prec[months_order]
```

```python
df_prec = df_prec.reset_index().rename_axis(None, axis=1)
```
