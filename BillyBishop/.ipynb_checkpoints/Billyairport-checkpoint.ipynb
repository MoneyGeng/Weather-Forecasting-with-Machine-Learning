```python
# Import findspark and initialize. 
import findspark
findspark.init()
```

```python
# Start Spark session
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("sparkFunctions").getOrCreate()
```

```python
from pyspark import SparkFiles
```

```python
options = {
    "header": "true",                    
    "ignoreLeadingWhiteSpace": "true"     
}
```

```python
df = spark.read \
    .option("header", options["header"]) \
    .option("ignoreLeadingWhiteSpace", options["ignoreLeadingWhiteSpace"]) \
    .csv("Resources/billy.csv")
```

```python
df.show()
```

```python
df.createOrReplaceTempView("my_table")
result = spark.sql("SELECT * FROM my_table")
```

```python
import pandas as pd
pandas_df = df.toPandas() 
```

```python
pandas_df
```

```python
pandas_df.dtypes
```

```python
pandas_df['Month'] = pandas_df['Month'].astype(int)
```

```python
pandas_df.dtypes
```

```python
new_temp_columns = ['Year','Month','Mean Temp (°C)', 'Date/Time']

# Display df with columns
df_temp = pandas_df[new_temp_columns]


df_temp
```

```python
# Create a dictory that maps that month names to the month numbers 
month_map = {1: 'JAN',
             2: 'FEB',
             3: 'MAR',
             4: 'APR',
             5: 'MAY',
             6: 'JUN',
             7: 'JUL',
             8: 'AUG',
             9: 'SEP',
             10:'OCT',
             11:'NOV',
             12:'DEC'}

df_temp['Month'] = df_temp['Month'].map(month_map)

df_temp
```

```python
# Drop duplicated rows
df_clean_temp = df_temp.drop_duplicates()

# Drop rows with missing values
df_clean_temp = df_temp.dropna()

#Drop index 
df_clean_temp.reset_index(drop= True)

df_clean_temp
```

```python
#year to int 
#mean temp to float
```

```python
df_clean_temp['Year'] = df_clean_temp['Year'].astype(int)
```

```python
df_clean_temp['Mean Temp (°C)'] = df_clean_temp['Mean Temp (°C)'].astype(float)
```

```python
df_clean_temp.dtypes
```

```python
#save output 
df_temp.to_csv("Resources/Cleaned_Billy.csv")
```
