```python
import pandas as pd 
```

```python
# read and open the csv file 
df = pd.read_csv('Woodbridge_Temp_Data.csv')
df
```

```python
#Create the mean temperature dataset for pearson airport
#Grab the columns needed ['Year','Month','Mean Temp (°C)']

new_temp_columns = ['Year','Month','Mean Temp (°C)']

# Display df with columns
df_temp = df[new_temp_columns]


df_temp
```

```python
# Create a dictory that maps that month names to the month numbers 
month_map = {1: 'JAN',
             2: 'FEB',
             3: 'MAR',
             4: 'APR',
             5: 'MAY',
             6: 'JUN',
             7: 'JUL',
             8: 'AUG',
             9: 'SEP',
             10:'OCT',
             11:'NOV',
             12:'DEC'}

df_temp['Month'] = df_temp['Month'].map(month_map)

df_temp
```

```python
# Pivot the dataframe
df_pivot = df_temp.pivot(index='Year', columns='Month', values='Mean Temp (°C)')

# Add the 'Year' column back to the DataFrame
df_pivot['Year'] = df_pivot.index

# Reorder the columns
columns_order = ['Year', 'JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']
df_transformed_temp = df_pivot[columns_order]

#Reset index 
df_transformed_temp = df_transformed_temp.reset_index(drop =True).rename_axis(None, axis =1)

# Print the transformed dataset
df_transformed_temp
```

```python
# Drop duplicated rows
df_clean_temp = df_transformed_temp.drop_duplicates()

# Drop rows with missing values
df_clean_temp = df_transformed_temp.dropna()

df_clean_temp
```

```python
#Save the transformed dataset to a csv file
df_clean_temp.to_csv('woodbridge_temp.csv', index = False)
```

```python
#finding the missing years data
min_year = df_clean_temp['Year'].min()
max_year = df_clean_temp['Year'].max()

# Create a range of years from min_year to max_year (inclusive)
all_years = range(int(min_year), int(max_year) + 1)

# Get the unique years from the 'Year' column
existing_years = df_clean_temp['Year'].unique()

# Find the missing years by subtracting the existing years from the complete range
missing_years = set(all_years) - set(existing_years)

# Print the missing years
print("Missing years:")
for year in missing_years:
    print(year)
```

```python
# Create total Precip 
#Grab the columns needed ['Year','Month','Total Precip (mm)']

new_precip_columns = ['Year','Month','Total Precip (mm)']

# Display df with columns
df_precip = df[new_precip_columns]


df_precip
```

```python
# Create a dictory that maps that month names to the month numbers 
month_map = {1: 'JAN',
             2: 'FEB',
             3: 'MAR',
             4: 'APR',
             5: 'MAY',
             6: 'JUN',
             7: 'JUL',
             8: 'AUG',
             9: 'SEP',
             10:'OCT',
             11:'NOV',
             12:'DEC'}

df_precip['Month'] = df_precip['Month'].map(month_map)

df_precip
```

```python
# Pivot the dataframe
df_pivot_precip = df_precip.pivot(index='Year', columns='Month', values='Total Precip (mm)')

# Add the 'Year' column back to the DataFrame
df_pivot_precip['Year'] = df_pivot_precip.index

# Reorder the columns
columns_order = ['Year', 'JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']
df_transformed_precip = df_pivot_precip[columns_order]

#Reset index 
df_transformed_precip = df_transformed_precip.reset_index(drop =True).rename_axis(None, axis =1)

# Print the transformed dataset
df_transformed_precip
```

```python
# Drop duplicated rows
df_clean_precip = df_transformed_precip.drop_duplicates()

# Drop rows with missing values
df_clean_precip = df_transformed_precip.dropna()

df_clean_precip
```

```python
#Save the transformed dataset to a csv file 
df_clean_precip.to_csv('woodbridge_precip.csv', index = False)
```

```python
# finding the missing years
min_year = df_clean_precip['Year'].min()
max_year = df_clean_precip['Year'].max()

# Create a range of years from min_year to max_year (inclusive)
all_years = range(int(min_year), int(max_year) + 1)

# Get the unique years from the 'Year' column
existing_years = df_clean_precip['Year'].unique()

# Find the missing years by subtracting the existing years from the complete range
missing_years = set(all_years) - set(existing_years)

# Print the missing years
print("Missing years:")
for year in missing_years:
    print(year)
```
